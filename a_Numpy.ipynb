### Indexing
- íŠ¹ì • ìœ„ì¹˜ì˜ ë°ì´í„°ë¥¼ ê°€ì ¸ì˜¤ëŠ” ê²ƒ

1. ìœ„ì¹˜ ì¸ë±ì‹±(Location indexing): ì „ë‹¬í•œ ìœ„ì¹˜(ì¸ë±ìŠ¤)ì˜ ê°’ í•œ ê°œ ì¶”ì¶œ

3. íŒ¬ì‹œ ì¸ë±ì‹±(Fancy indexing): listë¥¼ ì „ë‹¬í•˜ì—¬ í•œ ë²ˆì— ì—¬ëŸ¬ ìš”ì†Œë¥¼ ì¶”ì¶œ, íŠ¹ì • ìœ„ì¹˜ ê°’ë“¤ì„ ì½• ì°ì–´ì„œ ì¶”ì¶œ

```python
# 1-1. 1ì°¨ì› ìœ„ì¹˜ ì¸ë±ì‹±(Location indexing)
# 2ë¶€í„° 10ê¹Œì§€ ìˆœì„œëŒ€ë¡œ ìš”ì†Œë¥¼ ê°–ëŠ” 1ì°¨ì› ndarray
ndarray = np.arange(start=2, stop=11)
ndarray

data = ndarray[2]
print(data, type(data))

data = ndarray[-1]
print(data, type(data))


data = ndarray[-2]
print(data, type(data))


ndarray[-1] = 100
print(ndarray)
```

```python
# 1-2. 2ì°¨ì› ìœ„ì¹˜ ì¸ë±ì‹±(Location indexing)
ndarray1 = np.arange(start=1, stop=10)
ndarray2 = ndarray1.reshape((3, -1))
print(ndarray2)

print(ndarray2[1, 0])
print(ndarray2[1, 1])
print(ndarray2[2, 0])
print(ndarray2[2, 1])
print(ndarray2[0], ndarray2[0].shape)
```

```python
# 2-1. 1ì°¨ì› ìŠ¬ë¼ì´ì‹±(Slicing)

ndarray1 = np.arange(start=2, stop=10, step=2)
print(ndarray1)

print(ndarray1[:3], ndarray1[:3].shape)
print(ndarray1[3:], ndarray1[3:].shape)
print(ndarray1[:])
print(ndarray1[-4:])
print(ndarray1[-2::-1])
```

```python
# 2-2. 2ì°¨ì› ìŠ¬ë¼ì´ì‹±(Slicing)

ndarray1 = np.arange(start=1, stop=28)
print(ndarray1)

print("=" * 30)

ndarray2 = ndarray1.reshape((-1, 3))
print(ndarray2)

print("=" * 30)

print(ndarray2[:3, :2])

print("=" * 30)

print(ndarray2[4:9])

print("=" * 30)

print(ndarray2[4:, :])

print("=" * 30)

print(ndarray2[:])

print("=" * 30)

print(ndarray2[-2:-1, 0:2])

print("=" * 30)

print(ndarray2[::-1])

print("=" * 30)

print(ndarray2[::-1, ::-1])
```

```python
# 3. íŒ¬ì‹œ ì¸ë±ì‹±(Fancy indexing)

ndarray1 = np.arange(start=1, stop=21)
ndarray2 = ndarray1.reshape((4, 5))
print(ndarray2)

print(ndarray2[[0, 1, 3], 2:5])
print(ndarray2[[0, 1]])

# 2ì°¨ì› ndarrayì—ì„œ í–‰ê³¼ ì—´ ëª¨ë‘ íŒ¬ì‹œ ì¸ë±ì‹±ì„ ì‚¬ìš©í•˜ë©´, í•´ë‹¹ ì¢Œí‘œì˜ ì—´ë°±í„°ë¥¼ ê°€ì ¸ì˜¨ë‹¤.
# ndarray2[[x1, x2], [y1, y2]] : (x1, y1), (x2, y2) : [x, y]
print(ndarray2[[1, 2], [1, 3]])
print(ndarray2[[3, 2, 1], [1, 2, 4]])
```

```python
# 4. ë¶ˆë¦° ì¸ë±ì‹±(Boolean indexing)
ndarray1 = np.arange(start=1, stop=101, step=3)
print(ndarray1 % 5 == 0)
print(ndarray1[ndarray1 % 5 == 0])

ndarray1 = np.arange(3)
print(ndarray1)
print(ndarray1[[False, True, False]])
```

```python
# np.sort(ndarray)

original_ndarray = np.array([0, 4, 2, 5])

sorted_ndarray = np.sort(original_ndarray)
print(f"ì›ë³¸ ë°°ì—´: {original_ndarray}")
print(f"ì •ë ¬ëœ ë°°ì—´: {sorted_ndarray}")
```

```python
# np.sort(ndarray, axis=n)

ndarray1 = np.array([i for i in range(20, 0, -2)])
ndarray2 = ndarray1.reshape(2, -1)
print(f"ì›ë³¸ \n{ndarray2}")

ndarray2[0, 1] = 3
sorted_ndarray_axis0 = np.sort(ndarray2, axis=0)
print(f"axis0 ì •ë ¬ \n{sorted_ndarray_axis0}")

sorted_ndarray_axis1 = np.sort(ndarray2, axis=1)
print(f"axis1 ì •ë ¬ \n{sorted_ndarray_axis1}")
```

```python
# ndarray.sort()

original_ndarray = np.array([1, 2, 0, 4])
original_ndarray.sort()

print(f"ì›ë³¸ ë°°ì—´: {original_ndarray}")
print(f"ë‚´ë¦¼ ì°¨ìˆœ: {original_ndarray[::-1]}")
```

```python
# np.argsort(ndarray)

original_ndarray = np.array([0, 3, 2, 6])
sorted_indices = np.argsort(original_ndarray)
print(f'ì •ë ¬ í›„ ì›ë˜ ì¸ë±ìŠ¤: {sorted_indices}', type(sorted_indices))

sorted_indices_desc = sorted_indices[::-1]
print(f'ì •ë ¬ í›„ ì›ë˜ ì¸ë±ìŠ¤: {sorted_indices_desc}', type(sorted_indices_desc))
```

### ë‚´ì (Dot product)
- ë‘ ë²¡í„°ì˜ ì„±ë¶„ë“¤ì˜ ê³±ì˜ í•©.
- ë²¡í„°ì˜ ê¸¸ì´ ê³„ì‚°, ë‘ ê°œì˜ ë²¡í„°ê°€ ìˆ˜ì§ì¸ì§€, í‰í–‰í•˜ëŠ” ì§€, ë‘˜ ì‚¬ì´ì˜ ê°ì´ ì–¼ë§ˆì¸ì§€ ë“±ì˜ ì •ë³´ë¥¼ ì•Œì•„ë‚¼ ìˆ˜ ìˆë‹¤.

```python
# (n, m) * (n, m) = (n, m)
import numpy as np

ndarray1 = np.arange(1, 7).reshape((2, 3))
ndarray2 = np.arange(1, 7).reshape((2, 3)) + 2

print(ndarray1)
print(ndarray2)

ndarray3 = ndarray1 * ndarray2
print(ndarray3)
```

```python
# (n, m).dot((m, k)) = (n, k)

ndarray1 = np.arange(1, 7).reshape((2, 3))
ndarray2 = np.arange(1, 7).reshape((2, 3)) + 2

print(ndarray1)
print(ndarray2.T)

ndarray3 = ndarray1.dot(ndarray2.T)
print(ndarray3)

ndarray4 = ndarray1@ndarray2.T
print(ndarray4)
```

### ì„ í˜•ëŒ€ìˆ˜ì—ì„œ íˆ¬ì˜ì„ ì‚¬ìš©í•˜ëŠ” ì´ìœ
