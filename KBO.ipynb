```python
import pandas as pd
KBO_batter_full_df = pd.read_csv('./datasets/KBO_batter_full.csv')
# display(KBO_batter_full_df)

qualified_players = KBO_batter_full_df[KBO_batter_full_df['타수'] >= 5]

dusan_players = qualified_players[qualified_players['팀'] == '두산']


dusan_players['타율'] = pd.to_numeric(dusan_players['타율'], errors="coerce")


sorted_dusan_players = dusan_players.sort_values(by='타율', ascending=False)


top_10_dusan_players = sorted_dusan_players.head(10)


display(top_10_dusan_players)

```
