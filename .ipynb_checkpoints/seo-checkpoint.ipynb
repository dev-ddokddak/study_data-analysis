```python
customer_shopping_df['invoice_date'] = pd.to_datetime(customer_shopping_df['invoice_date'], dayfirst=True)
customer_shopping_df
```

```python
current_date = pd.Timestamp.today()

# 각 고객별로 최근의 구매 날짜
last_purchase = customer_shopping_df.groupby('customer_id')['invoice_date'].max()

# 현재 날짜와 최근의 구매 날짜의 차이
recency = (current_date - last_purchase).dt.days

print(recency)
```

```python
frequency = customer_shopping_df.groupby('customer_id')['invoice_no'].nunique()

frequency_df = frequency.reset_index()
frequency_df.columns = ['customer_id', 'Frequency']

print(frequency_df)
```

```python
customer_shopping_df['TotalSpend'] = customer_shopping_df['quantity'] * customer_shopping_df['price']
monetary = customer_shopping_df.groupby('customer_id')['TotalSpend'].sum()

monetary_df = monetary.reset_index()
monetary_df.columns = ['customer_id', 'Monetary']


customer_shopping_df
```
